```python
import pandas as pd
from folium.plugins import HeatMap
import folium
```

```python
df=pd.read_csv('dfMortesCompilado.csv')
df.head()
```

```python
df.info()
```

```python
df=df.drop(columns=['PROFISSAO','DESDOBRAMENTO','DESCR_PERIODO','FLAG_VITIMA_FATAL','CONDUTA','COORPORACAO','SITUACAO'])
```

```python
df = df.dropna()
```

```python
lat_lon=[-23.58633467412642, -46.60418260319519]
```

```python
df['CRIME'].unique()
```

```python
from folium import plugins
coordenadas=[]
for lat,lon in zip(df.LATITUDE[:3601],df.LONGITUDE[:3601]):
  coordenadas.append([lat,lon])

sp=folium.Map(location=[lat_lon[0],lat_lon[1]],zoom_start=10,tiles='cartodbpositron')
sp.add_child(plugins.HeatMap(coordenadas))
sp
```
